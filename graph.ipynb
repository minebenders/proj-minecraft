```python
import numpy as np
import networkx as nx
```

```python
A = np.zeros((8,8))

A[0][1] = 1
A[0][4] = 1
A[0][5] = 1
A[1][2] = 1
A[1][3] = 1
A[2][3] = 1
A[5][6] = 1
A[6][7] = 1
A[7][3] = 1
```

```python
G = nx.from_numpy_matrix(A, parallel_edges=False, create_using=nx.DiGraph)
```

```python
import matplotlib.pyplot as plt
```

```python
nx.draw(G, with_labels=True, font_weight='bold')
```

```python
nx.has_path(G, 4, 3)
```
